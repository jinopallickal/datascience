```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
file = pd.read_csv("Sub_Division_IMD_2017.csv")
```

```python
file
```

```python
Kerala = file[file.SUBDIVISION == 'Kerala']
```

```python
Kerala
```

```python
plt.plot(Kerala.YEAR, Kerala.ANNUAL/10)
plt.plot(Kerala.YEAR, Kerala.JJAS/10)
plt.title("Annual Rainfall for Kerala in cms")
plt.xlabel("YEARS")
plt.ylabel("Rainfall in cm")
plt.legend(["Annual", "JJAS"])
#plt.show()
plt.savefig('kerala.png')
```

```python
TamilNadu = file[file.SUBDIVISION == 'Tamil Nadu']
```

```python
TamilNadu
```

```python
plt.plot(TamilNadu.YEAR, TamilNadu.ANNUAL/10)
plt.title("Annual Rainfall for TamilNadu in cms")
plt.xlabel("YEARS")
plt.ylabel("Rainfall in cm")
#plt.show()
plt.savefig('TamilNadu.png')
```

```python
#Compare annual rainfall for year 1901-2017 for TamilNadu and Kerala

Kerala = file[file.SUBDIVISION == 'Kerala']
TamilNadu = file[file.SUBDIVISION == 'Tamil Nadu']

plt.plot(Kerala.YEAR, Kerala.ANNUAL/10)
plt.plot(TamilNadu.YEAR, TamilNadu.ANNUAL/10)
plt.title("Annual Rainfall for Kerala and TamilNadu in cms")
plt.xlabel("YEARS")
plt.ylabel("Rainfall in cm")
plt.legend(["Kerala","TamilNadu"])
#plt.show()
plt.savefig('comparision.pdf')
```
