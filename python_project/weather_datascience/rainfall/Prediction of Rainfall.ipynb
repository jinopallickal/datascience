```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("Sub_Division_IMD_2017.csv")
```

```python
file.info()
```

```python
#Total number of SubDivisions and Names of SubDivision

subdivs = file['SUBDIVISION'].unique().tolist()
```

```python
subdivs
```

```python
#Average Annual Rainfall in Each Subdivision

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
file.groupby('SUBDIVISION').mean().sort_values(by='ANNUAL', ascending=False)['ANNUAL'].plot(kind='bar', color='b',width=0.5,title='Subdivision wise Average Annual Rainfall', fontsize=25)
plt.xticks(rotation = 90)
plt.ylabel('Average Annual Rainfall (mm)')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
plt.show()
```

```python
print(file.groupby('SUBDIVISION').mean().sort_values(by='ANNUAL', ascending=False)['ANNUAL'])
```

```python
fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111)
file1 = file.groupby('YEAR').sum()['ANNUAL']
file1.plot(kind ='line', title='Overall Rainfall in Each Year', fontsize=20)
plt.ylabel('Overall Rainfall (mm)')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```python
#Monthly Rainfalls Plot (Subdivision wise)

months = file.columns[2:14]
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111)
xlbls = file['SUBDIVISION'].unique()
xlbls.sort()

file1 = file.groupby('SUBDIVISION').mean()[months]
file1.plot.line(title='Overall Rainfall in Each Month of Year', ax=ax,fontsize=20)
plt.xticks(np.linspace(0,35,36,endpoint=True),xlbls)
plt.xticks(  rotation = 90)
plt.ylabel('Rainfall (mm)')
plt.legend(loc='upper right', fontsize = 'xx-large')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```
