```python

import pandas as pd
import numpy as np

file = pd.read_csv("Sub_Division_IMD_2017.csv")
```

```python
file
```

```python
file.info()
```

```python
Kerala = file[file.SUBDIVISION == 'Kerala']
```

```python
Kerala
```

```python
x = Kerala.YEAR
```

```python
x
```

```python
y = Kerala.ANNUAL
```

```python
y
```

```python
z=Kerala.JJAS
u=Kerala.OND

plt.figure(figsize=(10,5))
plt.xlabel('YEAR')
plt.ylabel('Rainfall in mm')
plt.title('Trend of Rainfall in Kerala 1901-2017')
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,u)
plt.legend(['Annual','JJAS','OND'])
plt.savefig('kerala_rain.png')
plt.show()
```
